# Example of plotting with *matplotlib*

This notebook contains a very simple example of plot generation using the Python *matplotlib* package. The example is adapted from a script originally written in Python by Billy Armstrong in 2015.

Begin by importing *matplotlib*. Also run a "magic" command that allows plots to be displayed directly in the notebook. (Note: you need to have installed matplotlib).

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Define some data (here we're just making it up).

```python
x = [10, 50, 100, 500]  # list of values for the x-axis
y = [2, 10, 60, 90]     # list of values for the y-axis
```

There are a tons of plotting options. I've shown some of the ones I use most frequently here. Google "matplotlib plotting" to learn more than you've ever wanted to know.

```python
# Really basic version
plt.plot(x, y)
plt.xlabel('One thing [units]')      # This creates the x-axis label
plt.ylabel('Another thing [units]')  # This creates the y-axis label
```

```python
# Some slightly fancier options
# You can modify a lot of options using various optional keyword arguments
plt.plot(x, y, color='b', linewidth=2, linestyle='--', marker='x', markersize=10)
plt.xlabel('One thing [units]', fontsize=16)
plt.ylabel('Another thing [units]', fontsize=16)
```

Same thing but now adding another set of data, plus a title and legend:

```python
z = [3, 9, 56, 99]
plt.plot(x, y, color='b', linewidth=2, linestyle='--', marker='x', markersize=10)
plt.plot(x, z, color='r', linewidth=2, linestyle=':', marker='o', markersize=10)
plt.xlabel('One thing [units]', fontsize=16)
plt.ylabel('Another thing [units]', fontsize=16)
plt.title('One thing plotted against another', fontsize=18)
plt.legend(['Sample dataset 1', 'Sample dataset 2'], loc='upper left')
```
